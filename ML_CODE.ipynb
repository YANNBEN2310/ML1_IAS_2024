```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
 # Créer un modèle de régression linéaire
```

```python
def perform_linear_regression(x, y):
   
    model = LinearRegression()
    model.fit(x, y)

    # Générer des prédictions en utilisant le modèle
    y_pred = model.predict(x)

    return y_pred
```

```python
# Charger le jeu de données avec un délimiteur de point-virgule
try:
    data = pd.read_csv('your_dataset.csv', delimiter=';')

    # Prétraiter la colonne 'YearsExperience' en remplaçant les virgules par des points
    data['YearsExperience'] = data['YearsExperience'].str.replace(',', '.').astype(float)

    # Extraire les colonnes 'YearsExperience' et 'Salary'
    x = data['YearsExperience'].values.reshape(-1, 1)
    y = data['Salary'].values

    # Ajouter du bruit gaussien à la donnée 'Salary' avec une échelle de 100
    noise = np.random.normal(0, scale=100, size=y.shape)
    y_noisy = y + noise

    # Effectuer une régression linéaire
    np.random.seed(0)
    y_pred = perform_linear_regression(x, y_noisy)

    # Créer un graphique de dispersion
    plt.scatter(x, y_noisy, label='Données avec bruit')
    plt.plot(x, y_pred, color='red', label='Régression Linéaire')
    plt.xlabel('Années d\'expérience')
    plt.ylabel('Salaire')
    plt.title('Salaire vs Experience')
    plt.legend()
    plt.show()

except pd.errors.ParserError as e:
    print("Erreur de parsing: ", e)
```
